```python
import pandas as pd

df = pd.read_csv('tips.csv')

print('First 5 rows of the DataFrame:')
print(df.head())

print('\nInfo of the DataFrame:')
print(df.info())

print('\nAverage tip amount by sex:')
print(df.groupby('sex')['tip'].mean())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', data=df, palette='Set2')
plt.title('Total Bill Distribution by Day and Gender')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', hue='time', data=df)
plt.title('bigger total bill, bigger tip')
plt.show()
```
